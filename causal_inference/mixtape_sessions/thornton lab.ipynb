```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
df = pd.read_stata("https://raw.github.com/scunning1975/mixtape/master/thornton_hiv.dta", )
```

```python
df.head()
```

```python
# difference between the two means
df.groupby('any').agg({'got': 'mean'})
```

```python
df = df[(~df['got'].isna()) & (~df['any'].isna())]
```

```python
# ---- 1.b. Calculate using OLS
Y = df['got']
X = df['any']
# X = sm.add_constant(X)
model = sm.OLS(Y,X)
results = model.fit()
print(results.summary())
```
